# 1. What is linearity in linear regression

* It means that there is a linear relationship between the predictor variable (x) and the response variable (y).
* If the points in the plot fall roughly along a straight line, then the assumption is met.
* Linearity in linear regression is one of the four assumptions.
        1.Linear relationship
        2.Independence
        3.Normality
        4.Homoscedasticity:


# 2. Impact of outliers in machine learning model

* They can be caused by measurement errors, data entry errors
* Outliers can significantly affect the results of machine learning models, leading to inaccurate predictions   
* Causes overfitting or underfitting
 

# 3. Feature scaling techniques


*  It involves rescaling each feature such that it has a standard deviation of 1 and a mean of 0.
*  to ease the convergence (such as a non-penalized logistic regression), to create a completely different model fit compared to the fit with unscaled data (such as KNeighbors models).
#### Techniques:
 *       Absolute Maximum Scaling: 1.Find the absolute maximum value of the feature in the dataset , 2.Divide all the values in the column by that maximum value

*         Min-Max Scaling: n min-max you will subtract the minimum value in the dataset with all the values and then divide this by the range of the dataset(maximum-minimum).
*         Normalization                       
*         Standardization
*         Robust Scaling


# 4. What are the common problems with decision tree and how to overcome it


### Overfitting
* They can be unstable, meaning small changes in the data can lead to large changes in the structure of the tree and the predictions

### To overcome:
* Using pruning techniques, such as reducing the depth or size of the tree, to avoid overfitting and simplify the tree.
* Post pruning is to Find max depth then scale down
* Pre pruning
* Using ensemble methods, such as random forests or boosting
